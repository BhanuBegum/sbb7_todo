```python
import re

# Using re.compile()
pattern = re.compile(r'\b\d+\b')
text = "This year is 2021 and next year will be 2022."

# Using sub() to replace numbers with 'YEAR'
replaced_text = pattern.sub("YEAR", text)
print("Substituted Text:", replaced_text)

# Using split() to split text on numbers
split_text = pattern.split(text)
print("Split Text:", split_text)

# Demonstrating quantifiers
sample_text = "a aa aaa ab ac abc abcd"

# a|b : Matches 'a' or 'b'
print(re.findall(r'a|b', sample_text))

# + : Matches one or more occurrences
print(re.findall(r'a+', sample_text))

# * : Matches zero or more occurrences
print(re.findall(r'a*', sample_text))

# ? : Matches zero or one occurrence
print(re.findall(r'a?', sample_text))

# {p} : Matches exactly p times
print(re.findall(r'a{2}', sample_text))

# {p, q} : Matches between p and q times
print(re.findall(r'a{1,3}', sample_text))

# {p, } : Matches p or more times
print(re.findall(r'a{2,}', sample_text))

# {, q} : Matches up to q times
print(re.findall(r'a{,2}', sample_text))

```
